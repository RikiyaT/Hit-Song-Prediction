#combine 4 csv files

```python
import pandas as pd

# Load the CSV files into dataframes
df1 = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/lyrics0202.csv')
df2 = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/lyrics0209.csv')
df3 = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/lyrics0216.csv')
df4 = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/lyrics0223.csv')

# Concatenate the dataframes vertically
df_all = pd.concat([df1, df2, df3, df4], axis=0)

# Save the combined dataframe to a new CSV file
df_all.to_csv('all_files.csv', index=False)
```

#delete song data with no lyrics

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/all_files.csv')

# Drop the rows with no data in the third column
df.dropna(subset=[df.columns[2]], inplace=True)

# Write the new DataFrame to a new CSV file
df.to_csv('all_lyrics.csv', index=False)
```

#delete identical songs

```python
import pandas as pd

# read csv file
df = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/all_lyrics_neo.csv')

df.drop_duplicates(subset=['song'], keep='first', inplace=True)

# write to a new csv file
df.to_csv('/content/drive/MyDrive/siqi mock/csv files/all_lyrics_deleted.csv', index=False)
```

#delete words in the lyrics that are not needed

```python
import pandas as pd

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/all_lyrics_deleted.csv')

# Replace the text after "..." in the third column with an empty string
df['lyrics'] = df['lyrics'].str.replace(r'\.\.\..*', '', regex=True)

# Write the updated DataFrame to a new CSV file
df.to_csv('finaldata_all_lyrics.csv', index=False)
```

#combine senti data into song data

```python
import pandas as pd

# Read the first CSV file
df1 = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/all_us_songs.csv')

# Read the second CSV file
df2 = pd.read_csv('/content/drive/MyDrive/siqi mock/csv files/final_all_lyrics_negpos.csv')

# Merge the two dataframes based on 'track_name' and 'song'
merged_df = pd.merge(df1, df2[['song', 'bert_nega', 'bert_posi', 'bert_score', 'nltk_neg', 'nltk_neu', 'nltk_pos', 'nltk_score']], left_on='track_name', right_on='song', how='left')

# Drop the 'song' column
merged_df = merged_df.drop('song', axis=1)

merged_df
```

```python
# Save the merged dataframe to a new CSV file
merged_df.to_csv('merged_file.csv', index=False)
```
