```python
import pandas as pd
import glob

# Get a list of all CSV files in the current directory
file_list = glob.glob("/content/drive/MyDrive/siqi mock/csv files/regional-us-weekly-2023-02-02.csv") + glob.glob("/content/drive/MyDrive/siqi mock/csv files/regional-us-weekly-2023-02-09.csv") + glob.glob("/content/drive/MyDrive/siqi mock/csv files/regional-us-weekly-2023-02-16.csv") + glob.glob("/content/drive/MyDrive/siqi mock/csv files/regional-us-weekly-2023-02-23.csv")

# Create an empty list to store the data frames
dfs = []

# Iterate over each file in the file list
for file in file_list:
    # Read the CSV file into a data frame
    df = pd.read_csv(file)
    # Append the data frame to the list
    dfs.append(df)

# Concatenate the list of data frames into a single data frame
combined_df = pd.concat(dfs, ignore_index=True)

# Write the combined data frame to a new CSV file
combined_df.to_csv("ussongs.csv", index=False)
```
