```python
import csv
import requests

def get_lyrics(song, artist):
    url = f'https://api.lyrics.ovh/v1/{artist}/{song}'
    response = requests.get(url)
    if response.status_code == 200:
        lyrics = response.json()['lyrics']
        return lyrics
    else:
        return None

with open('/content/drive/MyDrive/siqi mock/csv files/allsongs.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)
    song_index = header.index('track_name')
    artist_index = header.index('artist_names')
    rows = []
    for row in reader:
        song_name = row[song_index]
        artist_name = row[artist_index]
        # check if song and artist name are in the correct format
        if ' ' not in song_name or ' ' not in artist_name:
            continue
        # search for the song on Lyrics.ovh API
        lyrics = get_lyrics(song_name, artist_name)
        # if lyrics are found, add them to the rows list
        if lyrics:
            rows.append([artist_name, song_name, lyrics])

# write the rows list to a new csv file
with open('all_lyrics.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Artist', 'Song', 'Lyrics'])
    for row in rows:
        writer.writerow(row)
```

```python
import csv
import requests

# Read the data from the CSV file
with open('/content/drive/MyDrive/siqi mock/csv files/allsongs.csv', mode='r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader) # Skip the header row
    rows = list(csv_reader)

# Search for lyrics for each song and save them to a new CSV file
with open('all_lyrics1.csv', mode='w', newline='') as csv_file:
    fieldnames = ['Artist', 'Song', 'Lyrics']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader() # Write the header row

    for row in rows:
        artist = row[2]
        song = row[3]

        # Search for the lyrics
        url = f"https://api.lyrics.ovh/v1/{artist}/{song}"
        response = requests.get(url)
        data = response.json()

        # Check if the lyrics were found
        if 'error' in data:
            print(f"No lyrics found for {song} by {artist}")
            continue

        # Check if the lyrics are in the original language
        if 'lyrics' not in data or 'translated' in data:
            print(f"Skipping {song} by {artist} because it's not in the original language")
            continue

        lyrics = data['lyrics']
        writer.writerow({'Artist': artist, 'Song': song, 'Lyrics': lyrics})
        print(f"Lyrics saved for {song} by {artist}")

print("Done")
```

```python
import csv
import requests

# Read the CSV file and get the song names and artist names
with open('/content/drive/MyDrive/siqi mock/csv files/regional-us-weekly-2023-02-23.csv', 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    header = next(csvreader)  # skip the header row
    song_names = []
    artist_names = []
    for row in csvreader:
        song_names.append(row[3])
        artist_names.append(row[2])

# Set up the API parameters
base_url = "https://api.musixmatch.com/ws/1.1/"
api_key = "428bb8986770c3c5182bbd45f9707512"
headers = {
    "Accept": "application/json",
    "Content-Type": "application/json",
}

# Search for the tracks
tracks = []
for i in range(len(song_names)):
    params = {
        "apikey": api_key,
        "q_track": song_names[i],
        "q_artist": artist_names[i],
        "s_track_rating": "desc",
        "f_has_lyrics": 1,
    }
    response = requests.get(base_url + "track.search", headers=headers, params=params).json()
    if response['message']['header']['status_code'] == 200:
        track_list = response['message']['body']['track_list']
        if len(track_list) > 0:
            # Get the track ID of the best matching track
            track_id = track_list[0]['track']['track_id']
            tracks.append(track_id)

# Get the lyrics for each track
lyrics = []
for track_id in tracks:
    params = {
        "apikey": api_key,
        "track_id": track_id,
    }
    response = requests.get(base_url + "track.lyrics.get", headers=headers, params=params).json()
    if response['message']['header']['status_code'] == 200:
        lyrics_body = response['message']['body']['lyrics']['lyrics_body']
        if "(Instrumental)" not in lyrics_body:
            lyrics.append(lyrics_body.replace("\n", " "))

# Write the results to a new CSV file
with open('lyrics.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['artist', 'song', 'lyrics'])
    for i in range(len(tracks)):
        csvwriter.writerow([artist_names[i], song_names[i], lyrics[i]])
```
